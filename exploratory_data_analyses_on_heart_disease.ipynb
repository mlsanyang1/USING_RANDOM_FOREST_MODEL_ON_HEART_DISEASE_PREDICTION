TRAINING THE DATA

```python
x = dataframe

y = dataframe["Age"]

y.unique
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

USING RANDOM FOREST MODEL

```python
print("The X_Train shape is", (x_train.shape))
print("The X_Test shape is", (x_test.shape))
print("The Y_Train shape is", (y_train.shape))
print("The Y_Test shape is",  (y_test.shape))
# print(x_test.shape)
# print(y_train.shape)
# print(y_test.shape)

```

TRAINING THE DATA

```python
classsifier = RandomForestClassifier(n_estimators = 5, criterion = 'entropy', random_state = 0)
classsifier.fit(x_train, y_train)
```

PREDICTING THE TEST RESULTS

```python
y_pred = classsifier.predict(x_test)
y_pred
```

CONFUSION MATRIX

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

SHOWING THE ACCURACY SCORE

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

TRYING DIFFERENT NUMBER OF TREES AND GINI CRITERIA

```python
classsifier = RandomForestClassifier(n_estimators = 7, criterion = 'gini', random_state = 0)
classsifier.fit(x_train, y_train)
```

```python
y_pred = classsifier.predict(x_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
